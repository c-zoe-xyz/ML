```python
df.head()
```

```python
prod_pyear = df.groupby("year").totalprod.mean().reset_index()
type(prod_pyear)
```

```python
select_column = prod_pyear["year"]
X = select_column.values.reshape(-1, 1)
```

```python
prod_pyear.head()
```

```python
y = prod_pyear["totalprod"]
type(y)
```

```python
plt.scatter(X, y)
plt.show
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
print(regr.coef_[0], regr.intercept_)
```

```python
y_pred = regr.predict(X)
```

```python
plt.plot(X, y_pred)
plt.scatter(X, y)
plt.show()
```

```python
future_honey = np.array(range(2013, 2051))
print(future_honey)
```

```python
future_honey = future_honey.reshape(-1,1)
print(future_honey)
```

```python
future_honey_pred = regr.predict(future_honey)
```

```python
plt.plot(X, y_pred)
plt.plot(future_honey, future_honey_pred)
plt.scatter(X, y)
plt.show()
```
